```python
# -*- coding: utf-8 -*-

import gibbslda
import pandas as pd

# load dataset
f = 'E:/word_embeding_TM/GLDAW_model/test_rus_docs.csv'
df = pd.read_csv(f, delimiter=';',decimal='.', encoding='ANSI')
texts = [[text for text in doc.split()] for doc in df['original data']]


gibbs = gibbslda.CGibbsLDA()


# put dataset into GLDA model
gibbs.SetInputDocs(texts)


print("Calculations...")
```

```python
# Possible values for the ldamethod parameter:
# "lda"
# "islda" - not yet implemented
# "granulate" - in this case the parameter granwnd must be specified
# "embeddings" - model with embeddings
# "out_dir="outcsv" -folder where phi matrices will be stored 
# "300_wiki_embeddings.txt" - embedding, for calculations

# Set model parameters
gibbs.SetParameters(alpha=0.1, beta=0.1, topics=6, niters=300, nitersave=100, ldamethod="embeddings", fixtopics = 0, granwnd = 0, out_dir="outcsv")
# If the parameter ldamethod="embeddings", you must call SetEmbeddings() and specify the path to the embeddings file and the desired number of words

gibbs.SetEmbeddings("300_wiki_embeddings.txt", 30)

gibbs.StartCalculation()

print("OK")
```
