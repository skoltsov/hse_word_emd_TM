```python
from gensim.models.coherencemodel import CoherenceModel
from utils import calc_topic_coherence as coher
from dataset import DocDataset
import torch
import numpy as np
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

# GSM

```python
tasknames = ['lenta_lem_1', 'lenta_lem_2', '20topicsnews_lem_1', '20topicsnews_lem_2']

for taskname in tasknames:
    fig1, ax1 = plt.subplots() 
    for index in range(1,4):
        renyi=[]
        for topics in range(2,50,2):
            model = torch.load('results/GSM/GSM_' + taskname + '_' + str(topics) +'_' + str(index))
            topic_words = model.get_topic_word_dist()
            num_top, num_word = topic_words.shape
    # 
            thresh = 1/num_word
            word_ratio = np.count_nonzero(topic_words>thresh)
            #print(word_ratio)
            sum_prob = topic_words[topic_words>thresh].sum()
            Sh = np.log(word_ratio/(num_word*num_top))
            En = -np.log(sum_prob/num_top)
            Fen = En - Sh*num_top
            Renyi = -(Fen/(1- num_top))
            renyi.append(Renyi)
        ax1.plot([2*i for i in range(1,25)], renyi)
    print(taskname)
    print('-------------')
    plt.show()
    print('-------------')
    #topic_words = pd.DataFrame(topic_words, columns=[vocab])
    #path = 'results/lenta/matrices/'+ emb + '_' + str(topics) + '_train_' + str(train) + '.csv'
    #topic_words.to_csv(path)# phi
        
```
