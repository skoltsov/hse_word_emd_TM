```python
import os
import pickle
import pandas as pd
import numpy as np
```

```python
# run
# указываем каталог, в котором лежат матрицы phi

results0 = 'E:/word_embeding_TM/Renyi_entropy/phi_matrix/'
results1 = 'E:/word_embeding_TM/Renyi_entropy/'
```

```python
all_Renyi = []
coherence = []
topics_num = []

for i in os.listdir(results0):
    print('file name: ', i)
    # загружаем модель
    df = pd.read_csv(results0+i ,  delimiter = ';', encoding='ANSI')
    print(list(df))
    
    del(df['Unnamed: 0'])
    topic_words = df.to_numpy()
    num_word, num_top = topic_words.shape
    print('topic number: ', num_top,' words number: ', num_word)
    
    thresh = 1/num_word
    word_ratio = np.count_nonzero(topic_words>thresh)
    sum_prob = topic_words[topic_words>thresh].sum()
    #print(word_ratio, ' ', sum_prob)
    
    Sh  = np.log(word_ratio/(num_word*num_top))
    En = -np.log(sum_prob/num_top)
    Fen = En - Sh*num_top
    Renyi = -Fen/(1- num_top)
    all_Renyi.append(Renyi)
    topics_num.append(num_top)
    print('model calculation is finished: ' + i)
    
    
# joint two columns abd save them to file
import csv
from itertools import zip_longest

d = [topics_num, all_Renyi]
export_data = zip_longest(*d, fillvalue = '')
with open(results1 + 'Renyi_entropy.csv', 'w', encoding="utf", newline='') as myfile:
      wr = csv.writer(myfile,  delimiter=';')
      wr.writerow(("Number of clusters","Renyi entropy"))
      wr.writerows(export_data)
myfile.close()
```
